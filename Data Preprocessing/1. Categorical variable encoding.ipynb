# Use One-Hot Encoding with scikit-learing

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oh_enc = OneHotEncoder(sparse_output=False,dtype="int64")   # oh_enc is object 
# for droping one columns use drop="first or last"
```

```python
oh_enc_npy_arr = oh_enc.fit_transform(tips_df[["sex","smoker","day","time"]])
oh_enc_npy_arr   # return 2-D numpy array
```

```python
dummy_df.keys()
```

```python
pd.DataFrame(oh_enc_npy_arr,columns=['sex_Female', 'sex_Male', 'smoker_No',
       'smoker_Yes', 'day_Fri', 'day_Sat', 'day_Sun', 'day_Thur',
       'time_Dinner', 'time_Lunch'])
```
